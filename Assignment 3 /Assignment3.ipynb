## 1- Read the data file: Balance.csv

```python
import pandas as pd
df = pd.read_csv('Balance.CSV')
df
```

## 2- Create a new column by the name of (Total Balance) which is the sum of the deposit and the balance.

```python
df['Total Balance'] = df['Balance'] + df['Deposite']
df
```

## 3- Add 3 per cent to the Balance Column.

```python
df['Balance'] = df['Balance'] * 0.03 + df['Balance']
df
```

## 4- Remove the Custumer_Name column.

```python
df.drop('Customer_Name', axis =1, inplace = True)
df
```

## 5- Rename the Customer_ID to become ID.

```python
df = df.rename(columns={'Customer_ID':'ID'})
df
```
